```python
import numpy as np 
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
res = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(res.content,'lxml')
table = soup.find_all('table')[0] 
df = pd.read_html(str(table))
df = df[0]
df.head()
#getting the data and read into pandas dataframe
```

```python
for i in np.arange(len(df))[::-1]:
    if(df.iloc[i,2]  == 'Not assigned'):
        if(df.iloc[i,1]  != 'Not assigned'):
            df.iloc[i,2] = df.iloc[i,1]
# cleaning the neighborhood that is not assigned where the borough is assigned
for i in np.arange(len(df))[::-1]:
    if(df.iloc[i,1]  == 'Not assigned'):
        df  = df.drop(i)
        i= i -1
# taking out the rows where Borough is unassigned
```

```python
postcode = df.Postcode.unique()
#getting the new list of postcode
neighborhoodlist = []
for j in np.arange(len(postcode)):
    neighborhoodlist.append([])
    for i in np.arange(len(df)):
        if(df.iloc[i,1]  != 'Not assigned'):
            if(df.iloc[i,2] !='Not assigned'):
                if (df.iloc[i,0] == postcode[j]):
                    neighborhoodlist[j].append(df.iloc[i,2])
#getting the new list of neighborhood                
borough = []                   
for j in np.arange(len(postcode)):
    for i in np.arange(len(df)):
        if(df.iloc[i,0] == postcode[j]):
            borough.append(df.iloc[i,1])
            break;
#getting the new list of borough                
```

```python
newdf = pd.DataFrame()
```

```python
newdf['Postcode']= postcode
```

```python
newdf['Borough']= borough
```

```python
newdf['Neighborhood'] = neighborhoodlist
```
